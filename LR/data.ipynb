```python
import numpy as np
import pandas as pd

np.random.seed(42)

n = 200  # number of points

# Single feature
X = np.random.uniform(0, 100, n)

# True linear relation
y = 4*X + 15 + np.random.normal(0, 10, n)  # y = 4x + 15 + noise

# DataFrame
data = pd.DataFrame({
    "X": X,
    "Y": y
})

data.to_csv("simple_linear_data.csv", index=False)
print("Dataset created: simple_linear_data.csv")
```
