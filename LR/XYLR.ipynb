**Linear Regression**


```python
X_mean=dataset['X'].mean()
Y_mean=dataset['Y'].mean()
print(X_mean)
print(Y_mean)
```

```python
# calculating slope
# ((x-xmean)(y-ymean)).sum()/((X-X_mean)**2).sum()
X=dataset['X']
Y=dataset['Y']
num=((X-X_mean)*(Y-Y_mean)).sum()
den=((X-X_mean)**2).sum()

slope=num/den

print(slope)
```

```python
# line became=> y=3.99x+c
# c=y_mean-m*x_mean
intercept=Y_mean-slope*X_mean
print(intercept)
```

```python
# line becomes=> y=3.9921x+16.052
# predicted value
x_i=X
y_hat=slope*x_i+intercept
y_hat.head()
```

```python
# errors
errors=Y-y_hat
# mse
mse = (errors**2).mean()
print(mse)
```

```python
R_square= 1-((Y-y_hat)**2).sum()/((Y-Y_mean)**2).sum()
print(R_square)
```
