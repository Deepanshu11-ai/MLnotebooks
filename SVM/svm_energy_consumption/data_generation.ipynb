```python
import numpy as np
import pandas as pd

np.random.seed(42)
n = 300

temperature = np.random.uniform(15, 40, n)       # Â°C
humidity = np.random.uniform(30, 90, n)          # %
usage_hours = np.random.uniform(1, 24, n)        # hours/day

# true relationship (unknown to model)
energy = (
    0.8 * temperature +
    0.5 * humidity +
    2.5 * usage_hours +
    np.random.normal(0, 5, n)
)

df = pd.DataFrame({
    "Temperature": temperature,
    "Humidity": humidity,
    "Usage_Hours": usage_hours,
    "Energy_Consumption": energy
})

df.head()
```

```python
# create data directory if not exists
import os
os.makedirs("data", exist_ok=True)

# save dataset
df.to_csv("data/energy_consumption.csv", index=False)

print("Dataset saved to data/energy_consumption.csv")
```
