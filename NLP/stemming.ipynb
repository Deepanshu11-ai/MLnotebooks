```python
"""reducig word to root form using nltk
trying to solve a classification problem on text data
classify positive review and negative reviews"""
```

```python
stemming_words = [
    "running", "runs", "runner", "ran",
    "easily", "fairness",
    "playing", "played", "plays",
    "studies", "studying",
    "connect", "connected", "connecting", "connections",
    "compute", "computed", "computing", "computation",
    "organization", "organizations", "organizing", "organizer",
    "happiness", "helpful", "usefulness",
    "development", "developer",
    "classification", "classify", "classifier",
    "learning", "learned", "learner"
]
```

```python
# porter stemmer
from nltk.stem import PorterStemmer
stemming=PorterStemmer()
```

```python
for word in stemming_words:
    print(f"{word}--->{stemming.stem(word)}")
```

```python
# regex stemmer
from nltk.stem import RegexpStemmer
reg_stemmer=RegexpStemmer('ing$|s$|ed$|ion$|er$|ly$|ity$|y$|al$|ful$')
for word in stemming_words:
    print(f"{word}--->{reg_stemmer.stem(word)}")
```

```python
# snowball stemmer
from nltk.stem import SnowballStemmer
snow_stemmer=SnowballStemmer('english')
for word in stemming_words:
    print(f"{word}--->{snow_stemmer.stem(word)}")
    
```
