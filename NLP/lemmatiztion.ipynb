```python
import nltk
from nltk import WordNetLemmatizer
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
words = ["running", "studies", "better", "playing", "feet"]
memmatized_words = [lemmatizer.lemmatize(word) for word in words]
print(memmatized_words)
    
```
