```python
sentence="The Eiffel Tower was built from 1887 to 1889 by Gustave Eiffel," 
" whose company specialized in building metal frameworks and structures."
```

```python
import nltk
from nltk import word_tokenize, pos_tag, ne_chunk
```

```python
nltk.word_tokenize(sentence)
```

```python
tag_elements=pos_tag(word_tokenize(sentence))
```

```python
import nltk

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('maxent_ne_chunker_tab')
nltk.download('words')
```

```python
named_entities=ne_chunk(tag_elements)
print(named_entities)
```
